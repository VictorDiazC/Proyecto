# **3.** Define una neurona que exhiba el mismo comportamiento, pero utilizando la clase `NeuronGroup` de la biblioteca Brian2.

```python
start_scope()

#params
tau = 35*ms
eqs = '''
dv/dt = (v0-v)/tau : volt (unless refractory)
v0 : volt
'''

#creamos 1 neurona
G = NeuronGroup(1, eqs,threshold='v > -40*mV',reset='v = -65*mV',refractory='2*ms',method='exact')

#'rest' de neurona en -70mV
G.v = -70*mV

#que no empiece inmediatamente
G.v0 = 0*mV

#creamos StateMonitor para registrar potencial de membrana 'v'
M = StateMonitor(G, 'v', record=True)

run(100*ms, report='text')

#grafica
plt.plot(M.t/ms, M.v[0])
plt.xlabel('Time (ms)')
plt.ylabel('v (volt)')

#línea de firing threshold:
plt.axhline(y=-0.040,linestyle='dashed', color='red')

plt.title('Membrane Potential vs. Time')
plt.grid(True)
plt.show()
```

En este caso, con un valor de tau = 35, se obviete una simulacion muy similar con 5 disparos al modelo de disparador con fugas del inciso 2.
